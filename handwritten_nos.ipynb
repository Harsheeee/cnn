```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test)=keras.datasets.mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
custom= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
custom.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
custom.fit(x_train, y_train, epochs=5)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred=custom.predict(x_test)
y_pred
```

```python
y_pred_ans= [np.argmax(i) for i in y_pred]
y_pred_ans[:5]
```

```python
print(classification_report(y_test, y_pred_ans))
```

```python
import seaborn as sns
cm=confusion_matrix(y_test, y_pred_ans)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
cnn=keras.Sequential([
    keras.layers.Conv2D(filters=10, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x_train, y_train, epochs=5)
```

```python
y_pred_cnn=cnn.predict(x_test)
y_pred_cnn
```

```python
y_pred_cnn_ans= [np.argmax(i) for i in y_pred_cnn]
y_pred_cnn_ans[:5]
```

```python
print(classification_report(y_test, y_pred_cnn_ans))
```

```python
cm_cnn=confusion_matrix(y_test, y_pred_ans)
sns.heatmap(cm_cnn, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
